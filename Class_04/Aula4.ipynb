## Plots Iniciais: Gr√°fico de Linhas üìà

Enfim, bora plotar uns gr√°ficos!

Vamos come√ßar com um exemplo simples. Para fazer qualquer gr√°fico, usamos o m√©todo `.plot` da nossa biblioteca.

Vamos entender o que aconteceu ali em cima.\

A pr√≥pria biblioteca vai se encarregar de ligar todos os pontos que criamos, fazendo a linha que mostramos acima.

### T√≠tulo

Podemos adicionar um t√≠tulo pro nosso gr√°fico com o m√©todo `title()`:

```python
plt.plot([1,2,3],[2,4,6])

plt.title('nosso gr√°fico')

plt.show()
```

### Nomea√ß√£o de Eixos

Podemos colocar nomes nos eixos $x$ e $y$ com os m√©todos `xlabel()` e `ylabel()`, respectivamente:

```python
plt.plot([1,2,3], [2,4,6])

plt.title('nosso gr√°fico!!!')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Vamos tentar plotar dois gr√°ficos juntos? O jeito mais f√°cil de fazer isso √© chamando o `.plot` duas vezes.

```python
x1 = [1,2,3,4,5]
y1 = [2,4,6,8,10]

x2 = [0,2,4,8,16]
y2 = [4,4,4,4,4]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.title('gr√°fico DUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Beleza! Plotamos dois gr√°ficos juntos. Mas t√° meio feia a nossa imagem de retorno, n√£o acha? Vamos tentar mudar isso.\
Usando o m√©todo `.figure()`, podemos alterar as configura√ß√µes da imagem do gr√°fico que o matplot nos retorna. Dentre os par√¢metros que podemos colocar dentro do `.figure()`, temos o *figsize* e o *dpi*. O *figsize* √© usado pra escolhermos as propor√ß√µes da nossa imagem. Colocando $(3,4)$ para o *figsize*, por exemplo, o gr√°fico vai retornar com uma propor√ß√£o 3x4.\
J√° o *dpi* serve para realmente mudarmos o tamanho do nosso gr√°fico. *dpi* indica o n√∫mero de pontos por polegada (*dots per inch*). Ent√£o, quanto maior o n√∫mero que colocarmos, maior ser√° o gr√°fico.\
Enfim, bora deixar nosso gr√°fico maior! Vamos colocar a propor√ß√£o 5x3 e o *dpi* em 150:

```python
plt.figure(figsize=(5,3), dpi=150)

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.title('gr√°fico DUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Beleza! Demos uma mudada e agora o nosso gr√°fico j√° t√° mais bonitinho. Mas vamos tentar mudar mais algumas coisas.

### Cor das Linhas

Podemos escolher a cor de cada linha do nosso gr√°fico passando o par√¢metro `color` dentro do ``plot``.

```python
plt.figure(figsize=(5,3), dpi=150)

plt.plot((np.random.randn(1000).cumsum()), color="blue")
plt.plot((np.random.randn(1000).cumsum()), color="magenta")

plt.title('gr√°fico DUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Podemos passar cores pelos seus c√≥digos HEX tamb√©m.

```python
plt.figure(figsize=(5,3), dpi=150)

plt.plot((np.random.randn(1000).cumsum()), color="#ae93f9")
plt.plot((np.random.randn(1000).cumsum()), color="#48f07b")

plt.title('gr√°fico DUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Pra ficar mais legal, ao inv√©s de plotar aquelas linhas que fizemos l√° em cima, coloquei a fun√ß√£o `randn` da biblioteca numpy pra gerar 1000 n√∫meros aleat√≥rios pra gente, junto com o m√©todo `cumsum`, que nos retorna a soma cumulativa dos elementos.

### Estilo da Linha

Podemos mudar o estilo da nossa linha usando o par√¢metro `linestyle` dentro do `plot`. Temos as seguintes op√ß√µes pra este par√¢metro: '-' que nos desenhar√° a linha s√≥lida, '--' que far√° a linha tra√ßada, '-.' far√° a linha com pontos e ':' s√≥ fara os pontos. Vamos ver todos juntos?

```python
plt.figure(figsize=(5,4), dpi=150)

plt.plot((np.random.randn(30).cumsum()), color="#ae93f9", linestyle='-')
plt.plot((np.random.randn(30).cumsum()), color="#48f07b", linestyle='--')
plt.plot((np.random.randn(30).cumsum()), color="#e7f07e", linestyle='-.')
plt.plot((np.random.randn(30).cumsum()), color="#ff79c6", linestyle=':')

plt.title('gr√°fico QU√ÅDRUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

### Marcadores

Ainda, podemos colocar marcadores nos nossos gr√°ficos com o par√¢metro `marker`. Existem dezenas de poss√≠veis valores que podemos colocar, voc√™ pode ver a lista [aqui](https://matplotlib.org/3.5.3/api/markers_api.html#module-matplotlib.markers). Neste notebook, vou mostrar os mais comuns: 
* '.' vai colocar pontos;
* 'o' vai colocar bolinhas;
* '^' adiciona um tri√¢ngulo pra cima;
* '*' adiciona uma estrela

```python
plt.figure(figsize=(5,4), dpi=150)

plt.plot((np.random.randn(30).cumsum()), color="#ae93f9", linestyle='-', marker='.')
plt.plot((np.random.randn(30).cumsum()), color="#48f07b", linestyle='--', marker='o')
plt.plot((np.random.randn(30).cumsum()), color="#e7f07e", linestyle='-.', marker='^')
plt.plot((np.random.randn(30).cumsum()), color="#ff79c6", linestyle=':', marker='*')

plt.title('gr√°fico QU√ÅDRUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

Ficou meio estranho, n√£o? Podemos mudar o tamanho dos marcadores com outro par√¢metro: o `markersize`.

```python
plt.figure(figsize=(5,4), dpi=150)

plt.plot((np.random.randn(30).cumsum()), color="#ae93f9", linestyle='-', marker='.', markersize=6)
plt.plot((np.random.randn(30).cumsum()), color="#48f07b", linestyle='--', marker='o', markersize=5)
plt.plot((np.random.randn(30).cumsum()), color="#e7f07e", linestyle='-.', marker='^', markersize=5)
plt.plot((np.random.randn(30).cumsum()), color="#ff79c6", linestyle=':', marker='*', markersize=5)

plt.title('gr√°fico QU√ÅDRUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.show()
```

### Legendas

Colocar legendas em nossos gr√°ficos √© outro elemento essencial pra uma boa leitura. Podemos colocar uma legenda com o m√©todo `legend()`. Pra que ele adicione as legendas corretamente, colocamos o par√¢metro `label` dentro de cada `plot` que fizemos, com o nome da linha.

```python
plt.figure(figsize=(5,4), dpi=150)

plt.plot((np.random.randn(30).cumsum()), color="#ae93f9", linestyle='-', marker='.', markersize=6, label='roxo')
plt.plot((np.random.randn(30).cumsum()), color="#48f07b", linestyle='--', marker='o', markersize=5, label='verde')
plt.plot((np.random.randn(30).cumsum()), color="#e7f07e", linestyle='-.', marker='^', markersize=5, label='amarelo')
plt.plot((np.random.randn(30).cumsum()), color="#ff79c6", linestyle=':', marker='*', markersize=5, label='rosa')

plt.title('gr√°fico QU√ÅDRUPLO')

plt.xlabel('eixo X')

plt.ylabel('eixo Y')

plt.legend()

plt.show()
```

## Tipos de Gr√°ficos

Como todos devem saber, existem diversos tipos de gr√°ficos, gr√°ficos de setores (popularmente conhecido como gr√°fico de pizzas), linha, barras, etc. Vamos tentar explorar um pouco dos mais usados.

### Gr√°ficos de Pizza üçï

Vamos tentar responder algumas perguntas usando nosso dataset. Qual √© a porcentagem total de popula√ß√£o que cada continente representa?

Esta √© uma pergunta que pode ser resolvida a partir de um gr√°fico de pizza. O gr√°fico do tipo pizza √© um dos gr√°ficos mais f√°ceis e simples de se entender e aplicar.

Ent√£o bora fazer um gr√°fico do tipo Pizza que mostra a porcentagem da popula√ß√£o de 2022 separado por continente.

A gente pode come√ßar criando um novo dataframe que agrupa os pa√≠ses pelo continente. Temos uma fun√ß√£o pr√≥pria do pandas que faz exatamente isso! üò± Essa fun√ß√£o √© a `groupby()`.

#### **groupby()**

O m√©todo `groupby()` serve para agruparmos dois datasets em um.\
Colocando o nome da coluna que queremos agrupar, vamos obter um novo dataframe em que o √≠ndice √© a coluna que escolhemos. Precisamos informar ao pandas o que queremos fazer com as outras colunas tamb√©m. Fazemos isso colocando um segundo m√©todo depois do ``groupby()``. Neste caso, queremos somar a popula√ß√£o de todos os pa√≠ses de cada continente. Ent√£o, usamos o m√©todo `sum()`.

```python
continentes = wrld_pop.groupby('Continent').sum()
continentes
```

Veja que lindo nosso novo dataframe! Perceba que agora temos um novo dataset que agrupou todos os pa√≠ses pelos seus continentes e somou todos os valores de dentro das colunas. Alguns dados somados ali dentro n√£o fazem sentido, ent√£o podemos retir√°-los.

```python
continentes.drop(['Rank', 'Density (per km¬≤)', 'Growth Rate'], axis=1, inplace=True)
continentes
```

Pronto! Agora nosso dataframe s√≥ tem informa√ß√µes √∫teis! üòé

Podemos criar um gr√°fico de pizza com o m√©todo `.pie()`. A pr√≥pria biblioteca matplotlib vai se encarregar de calcular o tamanho do nosso setor. Ent√£o podemos colocar n√∫meros brutos no m√©todo e ele vai calcular quanto que cada elemento representa do valor total. Ficou dif√≠cil de entender ? Vamos ver o `.pie()` trabalhando. üòå

Dentro do `pie()`, a gente coloca a coluna do nosso dataframe que queremos que seja contada. Para colocarmos as legendas, adicionamos o par√¢metro `labels` e colocamos os nomes que representam cada n√∫mero. Nesse caso, como o nome dos continentes √© o √≠ndice do nosso dataframe, colocamos continentes.index como valor do par√¢metro.


Por fim, podemos colocar o par√¢metro `autopct` pra que possamos escolher como nossa porcentagem aparece. Nesse caso, vou usar `'%.2f%%'`. O primeiro % √© para indicar que os pr√≥ximos caracteres ser√£o um valor. O .2f pede que o n√∫mero seja representado como um float com 2 d√≠gitos depois da v√≠rgula. Por fim, o %% serve para mostrar o sinal de porcentagem depois do n√∫mero. Sim, precisamos colocar o dois sinais (%%) ao inv√©s de 1, sen√£o o Python interpreta que deveria ter algum n√∫mero ali üòë.

```python
plt.figure(figsize=(5,4), dpi=150)
plt.pie(continentes['2022 Population'], labels=continentes.index, autopct='%.2f%%')
plt.show()
```

Perceba que colocamos a coluna 2022 Population ali no `.pie()` e ele nos retornou a porcentagem certinha de cada pa√≠s! üòÆ\
N√≥s passamos um valor inteiro e a biblioteca se deu o trabalho de somar e encontrar a fra√ß√£o que cada continente representa no total da popula√ß√£o. Legal n√©? üòù

O nosso dataset inclusive tem esse dado de percentual. Podemos verificar como o c√°lculo t√° certo usando aqueles valores:

```python
continentes['World Population Percentage']
```

Olha como os valores ficaram parecidos! ü§≠

### Gr√°ficos de Barras üìä

Podemos criar gr√°ficos de barra usando o m√©todo `bar()`. Colocamos como par√¢metros o eixo $x$ e $y$, respectivamente.

Se tentarmos mostrar a popula√ß√£o de cada continente no modo de barras, colocamos o √≠ndice do nosso dataframe continentes como eixo $x$ e a popula√ß√£o em 2022 no eixo $y$:

```python
plt.figure(figsize=(8,3), dpi=150)
plt.bar(continentes.index, continentes['2022 Population'])
plt.ylabel('Popula√ß√£o em Bilh√µes')
plt.grid(axis='y')
plt.show()
```

Bah, mas a grid t√° na frente das barras... isso t√° muito estranho. Podemos mudar isso passando o par√¢metro `zorder` no m√©todo `bar` pra colocar a coordenada Z dele acima do grid. Um valor de 2 para o `zorder` ser√° o suficiente:

```python
plt.figure(figsize=(8,3), dpi=150)
plt.bar(continentes.index, continentes['2022 Population'], zorder=2)
plt.ylabel('Popula√ß√£o em Bilh√µes')
plt.grid(axis='y')
plt.show()
```

O m√©todo `barh` nos d√° um gr√°fico de barras horizontal:

```python
plt.figure(figsize=(8,3), dpi=150)
plt.barh(continentes.index, continentes['2022 Population'], zorder=2)
plt.xlabel('Popula√ß√£o em Bilh√µes')
plt.grid(axis='x')
plt.show()
```

### Histogramas üìä

Um histograma √© um gr√°fico que agrupa n√∫meros em intervalos. Estes intervalos s√£o chamados de *bins*. Isso significa que o gr√°fico mostra **quantos** dados se encaixam em determinada caracter√≠stica.

#### Bins
*Bin* √© uma palavra do ingl√™s que significa um lugar usado para armazenar, como uma caixa, ba√∫, etc. Ent√£o, para nossos histogramas, cada *bin* √© uma 'caixa' que guarda os valores que est√£o dentro de um intervalo (as paredes da caixa). Veja um exemplo:


Como ser√° que √© a distribui√ß√£o de pa√≠ses com rela√ß√£o ao crescimento populacional? ü§î\
Essa √© uma pergunta que pode ser respondida com um histograma! Podemos criar histogramas com o m√©todo `hist()` do matplotlib. Como par√¢metros, passamos a coluna queremos que seja contada. Neste caso, queremos que conte os pa√≠ses pela taxa de crescimento (*Growth Rate*):

```python
plt.figure(figsize=(5,3), dpi=150)
plt.hist(wrld_pop['Growth Rate'], zorder=2)
plt.xlabel('Taxa de Crescimento Populacional')
plt.ylabel('N√∫mero de Pa√≠ses')
plt.grid(axis='y')
plt.show()
```

Beleza! Fizemos nosso gr√°fico. O que ele nos mostra que a maioria dos pa√≠ses tem uma taxa de crescimento por volta de 1. Mas ele ficou meio estranho, n√£o acha? ü§® As barras n√£o est√£o centralizadas, t√° meio feio... e ainda por cima ele t√° mostrando os valores muito amontoados... Vamos mudar isso. üòâ

Podemos escolher os intervalos dos nossos *bins* passando uma lista pro par√¢metro `bins` no m√©todo `hist()`. Isso vai mudar tamb√©m como o nosso eixo $x$ √© apresentado na imagem. Vamos escolher alguns novos valores pro eixo $x$.

```python
plt.figure(figsize=(5,3), dpi=150)
plt.hist(wrld_pop['Growth Rate'], bins=[0.98,0.99,1,1.01,1.02,1.03,1.04,1.05], zorder=2)
plt.xlabel('Taxa de Crescimento Populacional')
plt.ylabel('N√∫mero de Pa√≠ses')
plt.grid(axis='y', zorder=0)
plt.show()
```

Agora ele est√° mais leg√≠vel! Aqui escolhemos mostrar s√≥ os pa√≠ses que est√£o entre 0.98 e 1.05 de taxa de crescimento populacional.

Bora tentar fazer o mesmo gr√°fico de taxa de crescimento, mas desta vez s√≥ com os pa√≠ses da Europa. Vamos come√ßar criando um novo dataframe que s√≥ tenha os pa√≠ses da Europa. Podemos usar o nosso velho amigo `loc` pra isso:

## M√©todo `merge()`

Existe um m√©todo do pandas muito poderoso que serve para juntarmos dois dataframes em um. O `merge()` vai colocar as colunas de um dos dataframes no outro e, se os dois tiverem colunas em comum, elas v√£o se manter intactas como uma.

Vamos tentar usar o `merge()` com dois datasets: um que representa o GDP (*Gross domestic product*) de cada pa√≠s e o nosso dataframe das popula√ß√µes. O GDP √© chamado de Produto Interno Bruto (PIB) no Brasil.\
Enfim, vamos come√ßar lendo o csv do GDP:

```python
gdp=pd.read_csv('gdp_by_country.csv')
gdp.head()
```

Antes de tentarmos juntar os dataframes, vamos *dropar* algumas dessas colunas que n√£o vamos usar, j√° que nosso dataset t√° muito grande.

```python
lista_pra_dropar=['1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2012','2013','2014','2015','2016','2017','2018','2019']
gdp.drop(lista_pra_dropar, inplace=True, axis=1)
gdp.rename(columns={'Country': 'Country/Territory'}, inplace=True)
gdp = gdp.set_index('Country/Territory')
gdp.head()
```

Agora t√° mais bala. Bora usar o merge! ü§Ø\
Dentro do m√©todo `merge()`, colocamos o dataframe que queremos juntar e podemos colocar par√¢metro `on` ao lado. O valor que damos pro `on` tem que ser uma coluna que os dois dataframes tenham em comum. Assim, os nossos dataframes v√£o se juntar *nestas* colunas.

```python
wrld_info = wrld_pop.merge(gdp, on='Country/Territory')

wrld_info.head()
```

üò± Olha as colunas do gdp l√° no final! Funcionou perfeitamente! Mas tem um detalhe bem importante nessas colunas: os valores novos possuem **v√≠rgulas**. üò≥ Isso n√£o √© bom. Vamos ver qual √© o tipo de dado dessas colunas.

### M√©todo `info()`

Existe um m√©todo para verificarmos qual o tipo de dado que cada coluna possui de forma r√°pida. Este m√©todo √© o `info()`, da pr√≥pria biblioteca pandas:

```python
wrld_info.info()
```

Olha ali como as colunas 2020, 2021 e 2022 s√£o objetos... üò° Significa que os dados que est√£o ali n√£o s√£o *n√∫meros*, e sim qualquer outra coisa que n√£o √© um n√∫mero üòµ. Pra podermos usar esses dados direito, vamos ter que transformar essas informa√ß√µes em tipo *float*. 

Vamos dar uma lembrada na aula de limpeza de dados. L√°, usamos o m√©todo `replace()` em m√∫ltiplos n√∫meros ao mesmo tempo. Como o `replace` funciona em strings, a gente coloca o ``.str`` antes pra transformar o objeto da nossa coluna em string.

```python
wrld_info['2020']=wrld_info['2020'].str.replace(',', '')
wrld_info.head()
```

Veja que belezinha! Agora vamos fazer isso pras outras colunas!

```python
wrld_info['2021']=wrld_info['2021'].str.replace(',', '')
wrld_info['2022']=wrld_info['2022'].str.replace(',', '')
wrld_info.head()
```

Perfeito! üòÄ\
Mas ainda tem um passo que precisamos fazer. Apesar de termos tirado as v√≠rgulas, os n√∫meros ali dentro viraram strings. Precisamos transform√°-los em floats. Como ser√° que podemos fazer isso? ü§î

Olha s√≥ a tecnologia: existe um m√©todo que faz isso pra gente. ü§≠ü§≠

### `astype()`
O pandas possui um m√©todo chamado `astype()` que nos permite passar um dicion√°rio. Escolhemos as colunas que queremos como as chaves do dicion√°rio e o tipo de dado como os valores do dicion√°rio. Assim a gente n√£o precisa passar de dado em dado pra transformar em float! A√≠ fica f√°cil em! ü•±

```python
wrld_info=wrld_info.astype({'2020':'float', '2021':'float', '2022':'float'})
wrld_info.head()
```

Agora sim est√° perfeito!

Agora que arrumamos nosso dataframe, bora tentar usar nossos novos dados pra formar um gr√°fico.\
Ser√° que o GDP do Brasil cresceu entre 2020 e 2022??? ü§îü§î Vamos tentar descobrir com um gr√°fico de barras.\
Primeiro, selecionamos o Brasil:

```python
brasil = wrld_info.loc[wrld_info['Country/Territory'] == 'Brazil']
brasil
```

Boa! Agora a gente cria duas listas com *o que* mostraremos no gr√°fico: uma para o eixo $x$ e outra para o eixo $y$. Lembrando que o eixo $x$ vai mostrar os anos e o $y$ o GDP do Brasil em cada um desses anos.

```python
anos=[2020,2021,2022]
gdp_brasil = list(brasil.iloc[0,16:19].values)
gdp_brasil
```

Veja que usamos o m√©todo `list()` para transformar o conjunto de valores que pegamos do dataframe em uma lista. Selecionamos a partir do `iloc` a linha 0 e as colunas do intervalo 16-19 e pegamos os valores da s√©rie que o `iloc` nos retorna.

Beleza, agora tudo o que precisamos fazemos √© colocar as nossas duas listas no m√©todo `bar()`. Primeiro a nossa lista do eixo $x$ e depois a lista do eixo $y$. Bora colocar uma corzinha BRASILEIRA no nosso gr√°fico tamb√©m! üòÅ

```python
plt.figure(figsize=(6,4), dpi=150)
plt.bar(anos, gdp_brasil, color='#009739', zorder=2)
plt.ylabel('GDP do Brasil (milh√µes de US$)')
plt.grid(axis='y')
plt.show()
```

Uau, ficou muito bom! Mas tem um detalhe chato ali... tem esses valores n√£o inteiros nos anos, fica meio feio, n√£o acha? Vamos tentar mudar isso.

Podemos escolher os *ticks* que ser√£o mostrados nos nossos eixos usando os m√©todos `xticks` e `yticks`. O nosso gr√°fico em si n√£o vai mudar, s√≥ os n√∫meros que aparecem ali.

```python
plt.figure(figsize=(6,4), dpi=150)
plt.bar(anos, gdp_brasil, color='#009739', zorder=2)
plt.ylabel('GDP do Brasil (milh√µes de US$)')
plt.xticks([2020,2021,2022])
plt.yticks([0,500,1000,1500,2000])
plt.grid(axis='y')
plt.show()
```

ü§© Est√° incr√≠vel! Tirando a parte que o GDP do Brasil caiu desde 2020... üò£\
Enfim, o que importa √© o gr√°fico.

## `value_counts()`

Quando trabalhamos com um dataset n√£o num√©rico, √© muitas vezes necess√°rio usar outras ferramentas para criarmos nossos gr√°ficos. Nesta mesma aula, coloquei um dataset diferente sobre Pok√©mon. Este dataset nos traz algumas informa√ß√µes sobre cada pok√©mon desde a gera√ß√£o 1 at√© a 6.

```python
pokemon=pd.read_csv('Pokemon.csv')
```

```python
pokemon.head()
```

Usando o `.info()`, podemos ver os tipos de cada coluna:

```python
pokemon.info()
```

Como podemos perceber, temos diversas colunas que carregam dados de tipos n√£o num√©ricos. Suponha que a gente queira criar um gr√°fico de barras que mostre o n√∫mero de Pok√©mon que cada tipo possui. N√£o podemos simplesmente colocar nosso dataset no m√©todo `bar()` porque os dados de tipos n√£o s√£o num√©ricos... O que podemos fazer, ent√£o?
Podemos usar o m√©todo `value_counts()`, que vimos algumas aulas atr√°s para contar o n√∫mero de apari√ß√µes de cada dado!

```python
type1pkm=pokemon['Type 1'].value_counts()
type(type1pkm)
```

```python
type1pkm
```

O `value_counts()` nos retorna uma s√©rie. Ent√£o, para acessar apenas os valores da nossa s√©rie, selecionamos apenas os atributos *values*. Para a lista de nomes dos tipos, selecionamos o atributo *index* da s√©rie, porque na s√©rie que o `value_counts()` nos retorna, o √≠ndice carrega os nomes.

```python
plt.figure(figsize=(15,4), dpi=300)

plt.bar(type1pkm.index, type1pkm.values, zorder=2)
plt.grid(axis='y')
plt.show()
```

√â claro, este gr√°fico que fizemos considera apenas o tipo 1 dos Pok√©mon. E se quisermos adicionar os Pok√©mon do tipo 2 na contagem tamb√©m? üßê

Podemos come√ßar seguindo o mesmo m√©todo de `value_counts`, desta vez com a coluna 'Type 2':

```python
type2pkm=pokemon['Type 2'].value_counts()
type2pkm.head()
```

#### `add()`

Beleza! Temos duas s√©ries, uma com a contagem dos Pok√©mon e seus tipos prim√°rios e outra com os seus tipos secund√°rios. E agora? Precisamos de s√≥ uma s√©rie para podermos *plotar*! üòì\
Mais uma vez, temos um m√©todo para nos salvar! O `add()` vai somar os valores das duas s√©ries quando o √≠ndice deles coincidirem, o que √© exatamente o que queremos!

```python
alltypepkm=type1pkm.add(type2pkm)
alltypepkm
```

*Plotando* n√≥s obtemos:

```python
plt.figure(figsize=(15,6), dpi=300)

plt.bar(alltypepkm.index, alltypepkm.values, zorder=2)
plt.grid(axis = 'y')

plt.show()
```

Veja que belo gr√°fico! ü§ó

Usando essas mesmas propriedades, podemos fazer um gr√°fico de pizza que mostra a porcentagem de Pok√©mon que s√£o lend√°rios e quantos n√£o:

```python
plt.figure(figsize=(4,3), dpi=150)

plt.pie(pokemon['Legendary'].value_counts().values, labels=pokemon['Legendary'].value_counts().index, autopct='%.2f%%', labeldistance=1.15)

plt.show()
```
